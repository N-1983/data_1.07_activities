```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('regression_data.csv')
```

```python
def log_transfom_clean_(x):
    x = np.log10(x)
    if np.isfinite(x):
        return x
    else:
        return np.NAN 
data['IC3_log'] = list(map(log_transfom_clean_, data['IC3']))
data['IC3_log'] = data['IC3_log'].fillna(np.mean(data['IC3_log']))
sns.displot(data['IC3_log'])
plt.show()
```

```python
iqr = np.percentile(data['IC3_log'],75) - np.percentile(data['IC3_log'],25)
upper_limit = np.percentile(data['IC3_log'],75) + 3*iqr
lower_limit = np.percentile(data['IC3_log'],25) - 3*iqr
data = data[(data['IC3_log']>lower_limit) & (data['IC3_log']<upper_limit)]
sns.displot(data['IC3_log'])
plt.show()
```

```python
from scipy import stats
data['IC5_] = np.where(data['IC5']<=0,0,data=['IC5'])
mean = np.sum(data['IC5_'])/len(data[data['IC5_']>0])
data['IC5_'] = data['IC5_'].replace(0,mean)
data = data.drop(['IC5'], axis=1)
xt2, lmbda2 = stats.boxcox(data['IC5_'])
sns.displot(xt2)
plt.show()
```
